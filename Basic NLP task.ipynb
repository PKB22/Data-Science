### Basic NLP task in understanding the text and it's classfication
###  I have taken the dataset from the following link : [Link](https://www.gutenberg.org/files/69244/69244-0.txt)

```python
import nltk
```

```python
#Importing the text file from the given textbook from https://www.gutenberg.org/files/69244/69244-0.txt, that I stored on my local machine
with open('C:\\Users\\pavan\\OneDrive\\Desktop\\69244-0.txt', 'r', encoding='utf-8') as textfile:
    dataset = textfile.read().replace('\n', '')
```

```python
print(dataset)
```

```python
# convert sentences in words
from nltk import word_tokenize
token=word_tokenize(dataset)
print(token)
```

```python
len(token)
```

```python
#convert the words in lowercase
for i in range(len(token)):
    token[i]=token[i].lower()
```

```python
#calling stop words of english
from nltk.corpus import stopwords
from string import punctuation
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stopwords_en = set(stopwords.words('english')) 
#combining the punctuation with stopwords
stopwords_en_withoutpunc = stopwords_en.union(set(punctuation))

filtered_token=([word for word in token if word not in stopwords_en_withoutpunc])
```

```python
print(filtered_token)
```

```python
#Stemming and Lemmatization using the PorterStemmer() 
from nltk.stem import PorterStemmer
Porter_stemmer = PorterStemmer()  
dataset_stemmered = [Porter_stemmer.stem(word) for word in filtered_token]
```

```python
print(dataset_stemmered)
```

```python
len(dataset_stemmered)
```

```python
#check for the unique word counts after removing the stopwords punctuation and Porter_stemmer
Final_words= set(dataset_stemmered)
```
